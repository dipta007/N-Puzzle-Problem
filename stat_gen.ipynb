```python
import pickle
```

```python
N = 3
STAT_FILE = f'./out/stat_{N}.pickle'
OUTPUT_FILE = f'./data/{N*N-1}puzzle.out'
```

```python
with open(STAT_FILE, 'rb') as f:
    stat = pickle.load(f)
```

```python
with open(OUTPUT_FILE, 'r') as f:
    out = f.readlines()
    out = [line.strip() for line in out]
    out = out[:len(stat["IDA_star"]["hamming"]['time']['all'])]
```

```python
stat.keys()
```

```python
def get_stat(vals):
    if len(vals) == 0:
        return None
    return {
        'median': round(sorted(vals)[len(vals)//2], 2),
        'mean': round(sum(vals)/len(vals), 2),
        'min': round(min(vals), 2),
        'max': round(max(vals), 2),
        'std': round((sum([(x - sum(vals)/len(vals))**2 for x in vals]) / len(vals))**0.5, 2)
    }
```

```python
for algo in stat.keys():
    # print(algo, stat[algo].keys())
    for heuristic in stat[algo].keys():
        # print(heuristic, stat[algo][heuristic].keys())
        # print(stat[algo][heuristic]['cost']['found'])
        # print(stat[algo][heuristic]['cost']['not_found'])
        impossible = 0
        possible = []
        got_impossible = len(stat[algo][heuristic]['cost']['not_found'])
        got_possible = stat[algo][heuristic]['cost']['found']
        
        for v in out:
            if v == 'impossible':
                impossible += 1
            else:
                possible.append(int(v))
        
        # print(f"{algo} {heuristic} {impossible} {got_impossible} {possible} {got_possible}")
        print(algo, heuristic, impossible == got_impossible and len(possible) == len(got_possible) and possible == got_possible)
        print("Total Samples", len(out))
        print("Cost Found", get_stat(stat[algo][heuristic]['cost']['found']))
        print("Nodes Expanded Found", get_stat(stat[algo][heuristic]['nodes_expanded']['found']))
        print("Time Found", get_stat(stat[algo][heuristic]['time']['found']))
        # print("Cost Not Found", get_stat(stat[algo][heuristic]['cost']['not_found']))
        print("Nodes Expanded Not Found", get_stat(stat[algo][heuristic]['nodes_expanded']['not_found']))
        print("Time Not Found", get_stat(stat[algo][heuristic]['time']['not_found']))
        # print("Cost All", get_stat(stat[algo][heuristic]['cost']['all']))
        print("Nodes Expanded All", get_stat(stat[algo][heuristic]['nodes_expanded']['all']))
        print("Time All", get_stat(stat[algo][heuristic]['time']['all']))
        print()

```
